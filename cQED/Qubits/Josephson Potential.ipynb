```python
mu_0 = cs.mu_0                                     #vacuum susceptiblity
mu_B = cs.physical_constants['Bohr magneton'][0]   #Bohr magneton in SI units
k_b = cs.k                                         #K boltzmann
e = cs.e                                           #electron charge
hbar = cs.hbar
h = cs.h
phi_0 = cs.physical_constants['mag. flux quantum'][0]
Rq = h/e**2.
```

```python
def JPot(EJ, phi):
    return EJ*np.cos(phi)

def LPot(L, phi):
    return (phi*phi_0)**2./(2.*L)

def EJ(L):
    return (phi_0/(2.*np.pi))**2./L
```

```python
Lgeo = 1000*10**-9  #geometric inductance
Lk = 1*10**-9 #kinetic inductance
EJ1 = EJ(Lk)

print EJ1/h*10**-9
```

```python
phi = np.linspace(-10.*np.pi, 10.*np.pi, 400)
LK01 = np.linspace(1.,10,3)*10**-10
```

```python
figPot, axes = plt.subplots()

axes.plot(phi/np.pi, LPot(Lgeo, phi)/h*10**-9, label = r'$L_{{geo}} = {}\,\mathrm{{[nH]}}$'.format(Lgeo*10**9))
for L in LK01:
    axes.plot(phi/np.pi, JPot(EJ(L), phi)/h*10**-9, label = r'$L_{{K}} = {}\,\mathrm{{[nH]}}$'.format(L*10**9))
    axes.plot(phi/np.pi, (JPot(EJ(L), phi)+LPot(Lgeo, phi))/h*10**-9)

    
axes.set_xlabel(r'$\varphi/ \pi$')
axes.set_ylabel(r'$E_{\mathrm{Pot.}} / h\,\mathrm{GHz}$')

axes.legend(loc = 'upper center', bbox_to_anchor = (0.5,-0.25), ncol = 3)
```
